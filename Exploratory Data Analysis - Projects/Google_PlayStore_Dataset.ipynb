### Data Cleaning: 

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df1 = df.copy()
df1 = df1.drop(df1.index[10472])
```

```python
df1[~df1['Reviews'].str.isnumeric()]
```

```python
# converting to int
df1['Reviews'] = df1['Reviews'].astype(int)
df1.info()
```

```python
df1.Size.unique()
```

```python
df1['Size'] = df1['Size'].str.replace("M","000")
df1['Size'] = df1['Size'].str.replace("k","")
```

```python
df1['Size'] = df1['Size'].replace("Varies with device",np.nan)
df1['Size'] = df1['Size'].astype(float)
```

```python
df1.Size.unique()
```

```python
df1.info()
```

```python
df1['Installs'].unique()
```

```python
things_to_replace = ['+',',']

for items in things_to_replace:
    df1['Installs'] = df1['Installs'].str.replace(items,"")

df1.Installs.unique()
```

```python
df1['Installs'] = df1['Installs'].astype(int)
df1.info()
```

```python
df1.Price.unique()
```

```python
# df1[df1['Price'] == "Everyone"]
```

```python
df1['Price'] = df1['Price'].str.replace("$","0").astype(float)
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1['Last Updated'].unique()
```

```python
df1['Last Updated'] = pd.to_datetime(df1['Last Updated'])
df1['Day'] = df1['Last Updated'].dt.day
df1['Month'] = df1['Last Updated'].dt.month
df1['Year'] = df1['Last Updated'].dt.year
```

```python
df1.info()
```

```python
df1.Type.unique()
```

```python
df1['Current Ver'].unique()
```

```python
df1['Android Ver'].unique()
```

```python
df1.to_csv('data/google_cleaned.csv')
```
