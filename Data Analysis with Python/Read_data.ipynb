### Read Data From Different Sources :

```python
import pandas as pd
from io import StringIO
```

```python
data = '{"employee":"James", "age":43, "emailid":"james@gmail.com", "job_profile":[{"title1":"team lead","title2":"manager"}]}'
df = pd.read_json(StringIO(data))
```

```python
df
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient="records")
```
