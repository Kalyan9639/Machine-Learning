## <u>5-Number Summary and BoxPlot</u> 

### ⭐ Below are the 5-Number summary:
1. min
2. max
3. Q1 (25 percentile)
4. Q2 (50 percentile)
5. Q3 (75 percentile)

````python
import numpy as np
````

````python
lst_marks = np.random.randint(100,size=17).tolist()
lst_marks
````

````python
minimum,Q1,Q2,Q3,maximum = np.quantile(lst_marks,[0,0.25,0.5,0.75,1.0])
````

````python
minimum,maximum,Q1,Q2,Q3
````

#### ➡️ Now, we have to find out the lower and higher fence. Any data points, above the higher fence and below the lower fence are considered as outliers. For this, we need to calculate `IQR=Q3-Q1` and 

```
- Lower Fence = Q1- 1.5*IQR
- Higher Fence = Q3 + 1.5*IQR
```

````python
IQR = Q3-Q1
print(IQR)
````

````python
lower_fence = Q1 - 1.5*IQR
higher_fence = Q3 + 1.5*IQR
````

````python
lower_fence,higher_fence
````

#### ➡️ Boxplot automatically calculates lower and higher fences automatically 

````python
import seaborn as  sns

sns.boxplot(lst_marks)
````
