# Assignment 9: Structured Arrays

1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.
2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.


````python
# Part - 01
# structured array data type
dt = np.dtype([('name','U10'),('age', 'i4'),('weight', 'f4')])

i = np.array([('Alice',25,55.5),('Bob',30,85.0),('Charlie',22,68.2),("Don",28,75.0)],dtype=dt)
print("\n Structured Array:\n",i)

# sort the array by age
sorted_by_age = np.sort(i,order='age')
print("\n Structured Array sorted by age:\n",sorted_by_age)
````

````python
# Part - 02
dt = np.dtype([("x","i4"),("y","i4")])
i = np.array([(1,2),(3,4),(5,6),(7,8)],dtype=dt)
print("Structured Array:\n",i)

# calculating euclidean distance between each pair of points i.e., distance between 2 points
x1,y1 = i[0]['x'], i[0]['y']
x2,y2 = i[1]['x'], i[1]['y']
x3,y3 = i[2]['x'], i[2]['y']
x4,y4 = i[3]['x'], i[3]['y']

dist_12 = np.sqrt((x2-x1)**2 + (y2-y1)**2)
dist_23 = np.sqrt((x3-x2)**2 + (y3-y2)**2)
dist_34 = np.sqrt((x4-x3)**2 + (y4-y3)**2)
dist_41 = np.sqrt((x1-x4)**2 + (y1-y4)**2)

print(f"\n Euclidean Distances between points:\n Distance between point 1 and 2: {dist_12}\n Distance between point 2 and 3: {dist_23}\n Distance between point 3 and 4: {dist_34}\n Distance between point 4 and 1: {dist_41}")
````

````python
coords = np.column_stack((i['x'],i['y']))
print("after column stack:\n",coords)

dist_12 = np.linalg.norm(coords[1]-coords[0])
dist_23 = np.linalg.norm(coords[2]-coords[1])
dist_34 = np.linalg.norm(coords[3]-coords[2])
dist_41 = np.linalg.norm(coords[0]-coords[3])

print(f"\n Euclidean Distances between points using linalg.norm:\n Distance between point 1 and 2: {dist_12}\n Distance between point 2 and 3: {dist_23}\n Distance between point 3 and 4: {dist_34}\n Distance between point 4 and 1: {dist_41}")
````

2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.


````python
# Part - 01
j = np.random.randint(5,21,size=(4,4))
masked_array = np.ma.masked_greater(j,10)

print("Original Array:\n",j)
print("Masked Array:\n", masked_array)

unmasked_sum = np.sum(masked_array)
print(unmasked_sum)
````

````python
# Part - 02
j = np.random.randint(100,size=(3,3))
marr = np.ma.array(j)
marr[np.eye(marr.shape[0], dtype=bool)] = np.ma.masked

print("Original array:\n",j)
print("\nArray with masked diagonal:\n",marr)

unmasked_mean = np.mean(marr)
print("mean of unmasked elements: ",unmasked_mean)

# replace the masked elements with mean of unmasked elements
final = marr.filled(unmasked_mean)
print("\n Array final:\n",final)
````

# Remember !!

## 1. Fancy Indexing (Selection by Position)
You pass a list of integers to act as the indices you want to "pick out."
**Use case:** When you know the exact coordinates/positions of the data you want.

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50])

# Pass a list of indices: [pos 0, pos 4, pos 2] 
print(arr[[0, 4, 2]])  # Output: [10, 50, 30]
```

## 2. Boolean Indexing (Selection by Condition)
You pass a mask of True/False values. NumPy only returns the elements where the value is True.
**Use case:** When you want to filter data based on a mathematical rule (e.g., "all numbers > 25").

```python
arr = np.array([10, 20, 30, 40, 50])

# Create a condition (mask)
mask = arr > 25 
print(arr[mask])       # Output: [30, 40, 50]
```

### Summary Comparison

|Type  |	How it selects |	Input Example|
|---|---|---|
|Fancy | By Index/Position |	arr[[0, 3, 4]] 
|Boolean |	By Condition/Logic |	arr[arr % 2 == 0]
