## Handling Imbalanced Dataset:

1. Up Sampling
2. Down Sampling

```python
import pandas as pd
import numpy as np
```

```python
# Set the seed for reproduciblity
np.random.seed(123)

# Create a dataframe with 2 classes
n_samples = 1000
class_0_ratio = 0.9

n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
# CREATE MY DATASET WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature2': np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature2': np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
df.target.value_counts()
```

---

### 1. <u>Up Sampling</u>:

```python
df_minority = df[df.target == 1]
df_majority = df[df.target == 0]
```

```python
from sklearn.utils import resample

df_minority_upsample = resample(df_minority,replace=True # samples with replacements of the data points,
         n_samples=len(df_majority), random_state=42)
```

```python
df_minority_upsample.shape
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsample])
```

```python
df_upsampled.shape
```

```python
df_upsampled['target'].value_counts()
```

### 2. <u>Down Sampling</u>:

For this also, creating the same dataframe as upsampling

```python
# Set the seed for reproduciblity
np.random.seed(123)

# Create a dataframe with 2 classes
n_samples = 1000
class_0_ratio = 0.9

n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = n_samples - n_class_0

# CREATE MY DATASET WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature2': np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature2': np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})

df1 = pd.concat([class_0,class_1]).reset_index(drop=True)
df1.head()
```

```python
df1.target.value_counts()
```

```python
# downsampling
df_minority = df[df.target == 1]
df_majority = df[df.target == 0]
```

```python
df_majority_downsample = resample(df_majority,replace=False,
                                  n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsample.shape
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsample])
df_downsampled.head()
```
