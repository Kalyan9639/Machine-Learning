## SMOTE(Synthetic Minority Oversamping Technique)
SMOTE is a technique used in machine learning to address imbalanced dataset where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances Of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,n_redundant=0,weights=[0.9])
```

```python
x.shape,y.shape
```

```python
import pandas as pd

df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df = pd.concat([df1,df2],axis = 1)
df.head()
```

```python
df.target.value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset
oversample = SMOTE(random_state=42)
x,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
x.shape,y.shape
```

```python
len(y[y==0]),len(y[y==0])
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df_oversampled = pd.concat([df1,df2],axis=1)
df_oversampled.head()
```

```python
df_oversampled.shape
```

```python
df_oversampled.target.value_counts()
```

```python
plt.scatter(df_oversampled["f1"],df_oversampled['f2'],c=df_oversampled.target)
```
