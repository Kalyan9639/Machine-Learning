### Target Guided Ordinal Encoding:
Target guided ordinal encoding is a method for encoding categorical variables based on their relationship with a target variable. This means that each category is replaced with a numerical value derived from the mean or median of the target variable linked to that category.

For example, if you have a dataset with a categorical feature like "City" and a target variable like "Price," you would calculate the mean price for each city, then replace the city names with these mean values. This transformation can help improve the training of machine learning models by converting categorical data into numerical form.

```python
import pandas as pd

df = pd.DataFrame(
    {
        'city': ['ny','london','Tokyo','ny', 'Tokyo'],
        'price':[200,150,300,250,180]
    }
)
df.head()
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoding'] = df['city'].map(mean_price)
```

### Assignment:
In the dataset of tips obtained using `sns.load_dataset('tips')`, Get encoded time w.r.t total_bill

```python
import seaborn as sns

df = sns.load_dataset('tips') # get encoder for time w.r.t total_bill
df.head()
```

```python
df.isna().sum()
```

```python
sns.histplot(df['total_bill'],kde=True)
```

***This represents the data is normally distributed***

```python
mean_time = df.groupby('time')['total_bill'].mean().to_dict()
mean_time
```

```python
df['time_encoded'] = df['time'].map(mean_time)
```
