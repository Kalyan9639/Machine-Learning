```python
def fill_na(df, df_dict):
    for i in df.columns:
        if (df[i].isna().sum() != 0) and (df[i].dtype == 'float64'):
            # Compute mode for the column
            median = df[i].median()
            df[i] = df[i].fillna(median)

        else:
            if df[i].isna().sum() != 0:
                mode = df[i].mode()
                if not mode.empty:
                    df[i] = df[i].fillna(mode[0])
    return df
```

```python
df = fill_na(df, df_dict)
df.isna().sum()
```
