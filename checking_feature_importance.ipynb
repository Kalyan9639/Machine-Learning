```python
import joblib

model = joblib.load("rf_model_on_synthetic_data.joblib")
```

```python
# Example snippet for Random Forest
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Datasets/industry_realistic_pricing_dataset.csv")
x = df.drop(["Sales Volume","Product ID"],axis=1)
# Get feature importance
importances = model.feature_importances_
feature_names = x.columns
forest_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)

# Plot
fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
ax.set_title("Feature Importances (MDI)")
plt.show()
```
