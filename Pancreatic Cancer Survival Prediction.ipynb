```python
df3 = df2.copy()
```

```python
from sklearn.utils import resample
def upsample_data(df):
    # Helper function to upsample a given category
    def upsample_category(df, column_name):
        classes = [df[df[column_name] == category] for category in df[column_name].unique()]
        max_samples = max(len(cls) for cls in classes)
        upsampled_classes = [resample(cls, replace=True, n_samples=max_samples, random_state=123) for cls in classes]
        return pd.concat(upsampled_classes).reset_index(drop=True)
    
    # Upsample each category
    df_obesity = upsample_category(df[['Obesity']], 'Obesity')
    df_stage_at_diagnosis = upsample_category(df[['Stage_at_Diagnosis']], 'Stage_at_Diagnosis')
    df_alcohol_consumption = upsample_category(df[['Alcohol_Consumption']], 'Alcohol_Consumption')
    df_weight_loss = upsample_category(df[['Weight_Loss']], 'Weight_Loss')
    df_back_pain = upsample_category(df[['Back_Pain']], 'Back_Pain') 
    df_abdominal_discomfort = upsample_category(df[['Abdominal_Discomfort']], 'Abdominal_Discomfort') 
    df_jaundice = upsample_category(df[['Jaundice']], 'Jaundice') 

    # Combine all upsampled data
    df_balanced = pd.concat([df_obesity, df_stage_at_diagnosis, df_alcohol_consumption, df_weight_loss, df_back_pain,
                            df_abdominal_discomfort,df_jaundice], axis=1)
    df_balanced = df_balanced.loc[:,~df_balanced.columns.duplicated()]  # Remove duplicate columns

    return df_balanced
```

```python
df4 = upsample_data(df3)
```

```python
for i in df4.columns:
    print(df4[i].value_counts())
    print("-"*10)
```

```python
def upsample_data1(df):
    # Helper function to upsample a given category
    def upsample_category(df, column_name):
        classes = [df[df[column_name] == category] for category in df[column_name].unique()]
        max_samples = max(len(cls) for cls in classes)
        upsampled_classes = [resample(cls, replace=True, n_samples=max_samples, random_state=123) for cls in classes]
        return pd.concat(upsampled_classes).reset_index(drop=True)
    
    # Upsample each category
    df_hereditary_condition = upsample_category(df[['Hereditary_Condition']], 'Hereditary_Condition')
    df_family_history = upsample_category(df[['Family_History']], 'Family_History')
    df_chronic_pancreatitis = upsample_category(df[['Chronic_Pancreatitis']], 'Chronic_Pancreatitis')
    df_diabetes = upsample_category(df[['Diabetes']], 'Diabetes')
    df_smoking_history = upsample_category(df[['Smoking_History']], 'Smoking_History') 
    df_age = upsample_category(df[['Age']], 'Age') 
    df_survival_status = upsample_category(df[['Survival_Status']], 'Survival_Status') 

    # Combine all upsampled data
    df_balanced = pd.concat([df_hereditary_condition, df_family_history, df_chronic_pancreatitis, df_diabetes, df_smoking_history,
                            df_age,df_survival_status], axis=1)
    df_balanced = df_balanced.loc[:,~df_balanced.columns.duplicated()]  # Remove duplicate columns

    return df_balanced
```
