```python
df_balanced.columns
```

```python
df2 = df_balanced.copy()
df2.head()
```

```python
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
ohe = OneHotEncoder(drop = 'first')
```

```python
cnty = ohe.fit_transform(df2[['Country Name']]).toarray()
df_cnty = pd.DataFrame(cnty,columns = ohe.get_feature_names_out(['Country Name']))
df2 = pd.concat([df2,df_cnty],axis = 1)
df2.head()
```

```python
df3 = df2.drop("Country Name",axis = 1)
df3.head()
```

```python
df3.columns
```

```python
x = df3.drop("2023",axis = 1)
y = df3['2023']
```

```python
x = x.fillna(x.mean())
```

```python
y = y.fillna(y.median())
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled[:10]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.2,random_state = 42)
```

```python
x_train.shape,y_train.shape
```

# Linear Regression
---

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_model.score(x_test,y_test)
```

This may be because the pattern of the GDP is linear. so linear regression performed well

# Random Forest
---

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators = 10)
rf_model.fit(x_train,y_train)
```

```python
rf_model.score(x_test,y_test)
```

This might be the best to use because Random Forest is a non-linear model that tries to find patterns by making decisions based on feature splits

# Testing of data leakage
---

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

model_lr = LinearRegression()
scores = cross_val_score(model_lr, x_scaled, y, cv=5, scoring='r2')
print("Cross-validated R²:", scores.mean())  # If this is much lower than 0.99, there's overfitting!
```

```python
x_new = x.drop(columns=[str(year) for year in range(2015, 2023)])  
x_new_scaled = scaler.fit_transform(x_new)
x_train,x_test,y_train,y_test = train_test_split(x_new,y,test_size = 0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

This RandomForest model can likely make short-term predictions (like for the next year or two) relatively well,as long as the economic conditions remain stable and follow similar patterns to the past

# Visualizing

```python
import matplotlib.pyplot as plt

# Linear Regression prediction
y_pred_lr = lr_model.predict(x_test)

# Random Forest prediction
y_pred_rf = model.predict(x_test)

# Plot both models' predictions vs actual values
plt.figure(figsize=(12, 6))

# Linear Regression plot
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_lr)
plt.xlabel("Actual GDP in 2023")
plt.ylabel("Predicted GDP (Linear Regression)")
plt.title("Linear Regression Predictions")

# Random Forest plot
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_rf)
plt.xlabel("Actual GDP in 2023")
plt.ylabel("Predicted GDP (Random Forest)")
plt.title("Random Forest Predictions")

plt.tight_layout()
plt.show()
```
