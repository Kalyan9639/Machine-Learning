## Label Encoding:

Label Encoding and ordinal encoding are two techniques used to encode the category data as numerical data.

Label Encoding involves assigning a unique numerical label to each category in the variable. The labels are usually assigned in alphabetical order or based on the frequency of the categories. For example, if we have categorical variable color with 3 possible values (red, green, blue), it can be represented as:

- Red = 1
- Green = 2
- Blue = 3

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

le = LabelEncoder()
# create a simple dataframe
df = pd.DataFrame(
    {
        'color':['red','green','blue','green','red','blue']
    }
)
df.head()
```

```python
encoded = le.fit_transform(df[['color']])
encoded
```

```python
df_encoded = pd.concat([pd.DataFrame(encoded),df])
df_encoded.head()
```
