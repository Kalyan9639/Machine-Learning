### Data Visualization with Seaborn 

Seaborn is a python visualization library based on Matplotlib. So it is built on the top of matplotlib that provides a higher-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.Here, we will cover the basics to advance of Seaborn, including creating various types of plots and customising them

```python
import seaborn as sns
```

### Basic Plottings with Seaborn: 

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# create scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Line Plot
sns.lineplot(x='size',y='total_bill',data = tips,marker='o')
plt.title("Line Plot of Total Bill by Size")
plt.show()
```

### Categorical Plots :

```python
# bar plot
sns.barplot(x='day',y='total_bill',data = tips)
plt.title("Bar Plot of Total Bill by Day")
plt.show()
```

```python
# box plot
# this plot is very useful to identify outliers
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# violin plot
# this plot is similar to box plot but it also shows the distribution of the data

sns.violinplot(x='day',y='total_bill',data = tips)
```

```python
# Histograms

sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'],fill=True,color='g')
```

```python
# pair plot
sns.pairplot(tips)
```

```python
# heatmap

correlation = tips[['total_bill','tip','size']].corr()
correlation
```

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```

> ### You can visualize the grouped data using a **<u>bar plot</u>** or **<u>point plot</u>** in Seaborn.

*Here's an example:*

```python
tips.head()
```

```python
asdf = tips.groupby(['sex','day'])['tip'].mean().reset_index()
asdf
```

```python
sns.barplot(x='day', y='tip', hue='sex', data=asdf)
plt.title("Average Tip by Day and Sex (Bar Plot)")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```

```python
# point plot
sns.pointplot(x='day', y='tip', hue='sex', data=asdf, markers=['o', 's'], linestyles=['-', '--'])
plt.title("Average Tip by Day and Sex (Point Plot)")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```
